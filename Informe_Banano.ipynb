## **¿Cuál fue el país al que mas exporto banano Colombia durante el año 2020?**

#### A) Ecuador
#### B) España
#### C) Belgica
#### D) Estados Unidos

```python
url = 'https://raw.githubusercontent.com/BautistaDavid/Analysis-of-Banana-exports-in-Colombia/main/data/datos_exportaciones_banano.csv'
df = pd.read_csv(url)
```

```python
# five = df[(df['Country ID'] == 'eubel') |(df['Country ID'] == 'nausa') | (df['Country ID'] == 'eugbr') | (df['Country ID'] == 'euita') | (df['Country ID'] == 'eudeu')]
five_2020 = df[df['Year'] == 2020].sort_values('Trade Value', ascending = False).reset_index().loc[0:9]
five_2020['razon'] = (five_2020['Trade Value'] / 1074926842.0 ) *100
five_2020.drop(columns = ['index','Unnamed: 0'])
```

```python
rcParams['text.color'] = 'white'
rcParams['font.size'] = 15

sizes= five_2020['Trade Value']
label= five_2020['Country']
value = [str(round(i,2))+'%' for i in five_2020['razon']]
plt.figure(figsize = (12,7))
squarify.plot(sizes=sizes, label=label, alpha=0.8, value = value,pad=True)
plt.axis('off')
plt.text(50, 105, 'Top 10 países a los que Colombia\n exporto Bananas en 2020 ',fontsize = 22, color='Black', 
         horizontalalignment='center',verticalalignment='bottom',   zorder = 20)                 
plt.show()
```

```python
five = df[(df['Country ID'] == 'eubel') |(df['Country ID'] == 'nausa') | (df['Country ID'] == 'eugbr') | (df['Country ID'] == 'euita') | (df['Country ID'] == 'eudeu')]
```
