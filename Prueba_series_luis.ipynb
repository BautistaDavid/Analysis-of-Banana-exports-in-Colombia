### Por otro lado, algo que hay que tener en cuenta es que el dataframe ```expo_partners``` tiene datos de importaciones y exportaciones, por lo cual si se dan cuenta, el valor que aparecera para cada pais en cada año es un promedio entre exportaciones e importaciones.... Por ejemplo ... Francia 2001... dice que el valor es de 245339. Pero si ustedes se fijan al buscar en ```expo_partners``` las importaciones de francia para el 2001 seran de 350436, mientras que si buscan las importaciones el valor sera de 140242 ... de modo que: 

$$\frac{350436 + 140242}{2} = 245339$$

### Ahi la solucion facil es que filtren a ```expo_partners``` para que solo tenga exportaciones que creo que es como lo iban a hacer. 




````python
expo_partners.pivot_table(index='Años',columns='Paises',values='Valor')   # Pero esta pivot esta dando un promedio dentre importaciones y exportaciones de cada pais 
````
